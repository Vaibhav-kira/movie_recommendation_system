```python
data3['genres'] = data3['genres'].map(lambda x: ast.literal_eval(x))
data3['cast'] = data3['cast'].map(lambda x: ast.literal_eval(x))
data3['crew'] = data3['crew'].map(lambda x: ast.literal_eval(x))
```

```python
# next step is to take like actor_1_name , director_name from the dataset 
```

```python
def get_name(x):
  for i in x:
    print(i.get('name'),end=' ')
  print()
data3['genres'].head(5).map(lambda x : get_name(x))
```

```python
def make_genere_list(a):
  temp = []
  s = " "
  for i in a:
    if i.get('name') == 'Science Fiction':
      temp.append('Sci-Fi')
    else:
      temp.append(i.get('name'))
    if(temp == []):
      return np.nan
  return (s.join(temp))
```

```python
data3['genres_list'] = data3['genres'].map(lambda y: make_genere_list(y))
data3['genres_list'].head(5)
```

```python
def get_actor_1_name(a):
  casts = []
  for i in a:
    casts.append(i.get('name'))
  if casts == []:
    return np.nan
  else:
    return casts[0]
```

```python
data3['actor_1_name'] = data3['cast'].map(lambda y: get_actor_1_name(y))
data3['actor_1_name'].head(5)
```

```python
def get_actor_2_name(a):
  casts = []
  for i in a:
    casts.append(i.get('name'))
  if casts == [] or len(casts)<=1:
    return np.nan
  else:
    return casts[1]
```

```python
def get_actor_3_name(a):
  casts = []
  for i in a:
    casts.append(i.get('name'))
  if casts == [] or len(casts)<=2:
    return np.nan
  else:
    return casts[2]
```

```python
data3['actor_2_name'] = data3['cast'].map(lambda y: get_actor_2_name(y))
data3['actor_3_name'] = data3['cast'].map(lambda y: get_actor_3_name(y))
data3['actor_3_name'].head(5)
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data3['director_name'] = data3['crew'].map(lambda y: get_directors(y))
data3['director_name'].head(5)
```

```python
movie = data3[['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
```
